**B**: Aplicou alguma correção para o(s) problema(s) identificado(s) no item anterior?

1. Correções em ratings.csv (Outliers): Utilizando o método Interquartile Range (IQR) para identificar e tratar os outliers nas avaliações. O IQR é a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1) dos dados. Valores fora do intervalo [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR] são considerados outliers.

```python
import pandas as pd

ratings_df = pd.read_csv('/content/Database/ratings.csv')

Q1 = ratings_df['rating'].quantile(0.25)
Q3 = ratings_df['rating'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

ratings_df = ratings_df[(ratings_df['rating'] >= limite_inferior) & (ratings_df['rating'] <= limite_superior)]

ratings_df.to_csv('/content/Database/ratings_corrigido.csv', index=False)
```

2. Correções em links.csv (Dados faltantes): Remover as linhas com tmdbId faltantes, pois não podemos preenchê-las sem informações adicionais.

```python
import pandas as pd

links_df = pd.read_csv('/content/Database/links.csv')

links_df = links_df.dropna(subset=['tmdbId'])

links_df.to_csv('/content/Database/links_corrigido.csv', index=False)
```

3. Correções em tags.csv: Remover tags com baixa frequência, pois podem ser muito específicas ou irrelevantes. Definimos um limite mínimo de 5 ocorrências para uma tag ser considerada relevante.

```python
import pandas as pd

tags_df = pd.read_csv('/content/Database/tags.csv')

tag_counts = tags_df['tag'].value_counts()

tags_df = tags_df[tags_df['tag'].isin(tag_counts[tag_counts >= 5].index)]

tags_df.to_csv('/content/Database/tags_corrigido.csv', index=False)
```
